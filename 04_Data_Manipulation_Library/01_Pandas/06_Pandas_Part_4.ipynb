```python
## Time Delta
```

```python
pd.Timedelta('3 days 22 hours 30 minutes 15 seconds')
```

```python
td = pd.Timedelta('3 days 22 hours 30 minutes 15 seconds')

print(f"Days: {td.days}")
print(f"Seconds: {td.seconds}")
print(f"microseconds: {td.microseconds}")
print(f"nanoseconds: {td.nanoseconds}")
print(f"Total seconds: {td.total_seconds()}")
```

```python
td = pd.Timedelta('3 days 22 hours 30 minutes 15 seconds')
print(f"Components: {td.components}")
```
