```python
from bokeh.plotting import figure, show, output_file
from bokeh.models import ColumnDataSource
from bokeh.sampledata.iris import flowers
from bokeh.transform import factor_cmap
from bokeh.palettes import Category10

output_file("iris.html")
p = figure(title='test flower')
p.xaxis.axis_label = 'Petal Length'
p.yaxis.axis_label = 'Petal Width'

source = ColumnDataSource(flowers)
species = flowers['species'].unique()
colors = Category10[len(species)]

p.scatter(x="petal_length", y="petal_width",
          source=source,
          fill_alpha=0.2, size=10,
          legend_field="species",
          color=factor_cmap('species', palette=colors, factors=species))

show(p)
```

```python
x = [2, 4, 5, 6, 7, 8, 9]
y = [4, 3, 6, 7, 5, 7, 5]
```

```python
# Create a new figure for the line plot
line_p = figure(title="Line Plot")

# Add a line glyph to the figure
line_p.line(x, y, line_width=2)

# Show the plot
show(line_p)
```
