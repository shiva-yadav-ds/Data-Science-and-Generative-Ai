```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width, hue=iris.species)
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.petal_length, hue=iris.species)
```

```python
sns.scatterplot(x=iris.petal_length, y=iris.sepal_width, hue=iris.species)
```

```python
sns.displot(iris['sepal_length'])
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
tips['smoker'].value_counts()
```

```python
tips['sex'].value_counts()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='smoker', style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', size='smoker', hue='size')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, style='smoker', size='smoker', hue='size', col='sex')
```

```python
sns.catplot(x='day', y='total_bill', data=tips, kind='swarm')
```

```python
sns.relplot(x='smoker', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.pairplot(tips)
```
